```python
import pandas
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
################# 
# define color mapping
#################
```

```python
#####################################
# create master file for specific run
#####################################

run_folder = 'run_01'

if(not os.path.exists(f'../data/{run_folder}/Graphs')):
    os.mkdir(f'../data/{run_folder}/Graphs')


result_table_names = os.listdir(f'../data/{run_folder}/Results')

tables = []

for _table in result_table_names:
    _df = pandas.read_csv(f'../data/{run_folder}/Results/{_table}',sep=",",header=None)
    _df['Normalization'] = _table.split('_')[1].split('.')[0]
    tables.append(_df)

master_df = pandas.concat(tables,axis=0)
# rename columns
# '{Encoding}_{classifier}, AUC, ACC, Recall, Precision, F1_score, time-end, runtime(sec), classfication time(sec), best hyper-parameter'
master_df = master_df.rename(columns={0:'{Encoding}_{classifier}',1:'AUC',2:'ACC',3:'Recall',4:'Precision',5:'F1_score',6:'time-end',7:'runtime(sec)',8:'classfication time(sec)',9:'best hyper-parameter'})
master_df
```

```python
# plot overview over recall and precision for all models
colorDef = {'relative abundance':'b','softmax':'g','sigmoid':'r','TMM':'c'}

fig,ax = plt.subplots(1,2)
ax[0].set_title("Precision and Recall")
ax[1].set_title("Average Precision and Recall")
for normalization in set(master_df["Normalization"].to_list()):
    df = master_df[master_df["Normalization"] == normalization]
    df.plot.scatter(x="Recall",y="Precision",color=colorDef[normalization],label=normalization, ax=ax[0])
    ax[1].scatter(x=np.mean(df['Recall']),y=np.mean(df['Precision']),color=colorDef[normalization],label=normalization)
fig.savefig((f'../data/{run_folder}/Graphs/RecallAndPrecision'))
```

```python
# plot overview over relation between model and normalization 

_possibleScores = ['AUC', 'ACC', 'Recall', 'Precision', 'F1_score']
fig,ax = plt.subplots(5,figsize=(10,20))
norm = master_df["Normalization"].tolist()
_r = []
for i in range(len(norm)):
    if(norm[i] not in _r):
        _r.append(norm[i])
norm = _r

method = master_df["{Encoding}_{classifier}"].tolist()
_r = []
for i in range(len(method)):
    if(method[i] not in _r):
        _r.append(method[i])
method = _r

for s in range(len(_possibleScores)):

    scores = np.array(master_df[_possibleScores[s]].tolist()).reshape(4,6)


    im = ax[s].imshow(scores)
    ax[s].set_yticks(np.arange(len(norm)),labels=norm)
    ax[s].set_xticks(np.arange(len(method)),labels=method)

    plt.setp(ax[s].get_xticklabels(), rotation=45, ha="right",rotation_mode="anchor")
    ax[s].figure.colorbar(im,ax=ax[s])
    for x in range(len(norm)):
        for y in range(len(method)):
            ax[s].text(y,x,scores[x,y],ha="center",va="center", color="w")
    ax[s].set_title(f"{_possibleScores[s]}")
    fig.suptitle(f"Performance scores for method and normalization")
    fig.tight_layout(pad=5.0)
    fig.savefig((f'../data/{run_folder}/Graphs/Performance Scores'))
```
