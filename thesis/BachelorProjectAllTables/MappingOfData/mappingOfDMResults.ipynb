```python
import pandas as pd
import os
import seaborn as sns
import numpy as np
```

```python
dfs = []
for fileName in os.listdir('Results'):
    df = pd.read_csv(f"Results\\{fileName}")
    df.columns = '{Encoding}_{classifier}, AUC, ACC, Recall, Precision, F1_score, time-end, runtime(sec), classfication time(sec), best hyper-parameter'.replace(' ','').split(',')
    df['NormalizationMethod'] = fileName.split('.')[0].split('_')[1]
    df['TaxonomicRank'] = fileName.split('.')[0].split('_')[0]
    dfs.append(df)
master_df = pd.concat(dfs)
master_df
master_df.to_csv("master_df",sep="\t",index=False)
```

```python
#norm_tax = master_df.pivot(index="TaxonomicRank", columns="NormalizationMethod", values="F1_score")
#sns.heatmap(norm_tax)
# plotting combination of Normalization and Taxonomic Rank Against each other
norm_tax = master_df.filter(["F1_score","TaxonomicRank","NormalizationMethod"],axis=1)
#norm_tax
#sns.heatmap(norm_tax)
norm_tax.groupby(['NormalizationMethod',"TaxonomicRank"]).mean()
mapdf = pd.DataFrame()
mapdf.index = set(norm_tax["TaxonomicRank"])
for i in set(norm_tax["NormalizationMethod"]):
    mapdf[i] = np.nan
#mapdf.columns = set(norm_tax["NormalizationMethod"])
for index, row in norm_tax.iterrows():
    mapdf[row["NormalizationMethod"]][row["TaxonomicRank"]] = row["F1_score"] # 
    #print(row["NormalizationMethod"], row["TaxonomicRank"], row["F1_score"])
mapdf
```
